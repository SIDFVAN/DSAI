Suppose Ïƒ is unknown, calculate a 95% confidence interval for the population mean. (Answer = [487.319, 512.993])

Finally, suppose the sample consists of only the first 25 observations in this dataset. Calculate a 95% confidence interval for this situation.  
(Answer = [450.291,536.669])

```python
import math
# Step 1.
m = 324.6      # Sample mean =onbekend dus berekenen
m = df['Money'][:25].mean()
s = 0      # sigma is voor de standaard afwijking unknown , std afwijking van sample Sigma_sample
sigma_sample = df['Money'][:25].std()
n = 45      # bereken
n = df['Money'][:25].count()
alpha = .05  # 1 - alpha is the confidence level

# Step 2.
z = stats.t.isf(alpha/2,loc=0,scale=1,df=n-1)
print("z-score: %.5f" % z)

# Step 3.
lo = m - z * sigma_sample / np.sqrt(n)
hi = m + z * sigma_sample / np.sqrt(n)
print("Confidence interval: [%.4f, %.4f]" % (lo, hi))
```
